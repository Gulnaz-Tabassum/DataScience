2. Cyy=Covariance matrices between datasets y and y

```python
row,col=(n,n)
W=[]
for i in range(col):
    c=[]
    for j in range(row):
        if i==j:
            c.append(1/n)
        else:
            c.append(0)
    W.append(c)
W
```

### Calculating sigma : σ ← tr(Stx )/tr(Sty )


```python
sigma1= Stx.trace()
sigma2= Sty.trace()
sigma= sigma1/sigma2
sigma
```

Trace function above determines the trace of the matrix Stx, Sty 

```python
r,c=Stx.shape
```

Iu below is an diagonal matrix of size (r.c) i.e. (10*10) in this case.

```python
Iu=[]

for i in range(c):
    c1=[]
    for j in range(r):
        if i==j:
            c1.append(1)
        else:
            c1.append(0)
    Iu.append(c1)
Iu
```

Notations used in below part of the code will be

1.) Dx,Dy= Empty Matrices of shape (r,c) 

```python
d = Stx.shape[0]

Dx = [ [ 0 for i in range(r) ] for j in range(c) ]
Dy = [ [ 0 for i in range(r) ] for j in range(c) ]
```

### Calculation of matrix Px,Py can be done using following formula 



Px = I−StxDx(Transpose)*(DxStxDx(Transpose)^(-1))*Dx

Py = I−StyDy(Transpose)*(DyStyDy(Transpose)^(-1))*Dy



### Calculation of rth vector pair wxr,wyr 

The value of wxr,wyr can be find by each time in loop running through variable r by finding eigen value of matrix wx 
 
and taking maximum of all values coming everytime through loop

### Finding value of matrix Dx,Dy

The value of Dx and Dy can be find by appending wxr to Dx and wyr to Dy as the last column

```python

for u in range(0,d):
    #calculating Px
    p=np.dot(Stx,np.transpose(Dx))
    p1=np.dot((np.dot(Dx,Stx)),np.transpose(Dx))
    p1_inv=np.linalg.pinv(p1)

    Px=np.dot((np.dot(p,p1_inv)),Dx)
    Px = Iu - Px
    #Calculating Py
    P=np.dot(Sty,np.transpose(Dy))
    P1=np.dot((np.dot(Dy,Sty)),np.transpose(Dy))
    P1_inv=np.linalg.pinv(P1)

    Py=np.dot((np.dot(P,P1_inv)),Dy)
    Py= Iu-Py
    
    #Wx and Wy
    A = Sty*Px*Sbx
    A = sigma*A
    B = Sty*Px*Cxy
    B = sigma*B
    C = Stx*Py*Cxy
    D =Stx*Py*Sby
    
    Inv_a=np.linalg.pinv(A)
    Inv_d=np.linalg.pinv(D)

    F1=np.dot(Inv_a,B)
    F2=np.dot(F1,Inv_d)
    F=np.dot(F2,C)
    F[F<0]=0
    F_Final=np.sqrt(F)

    U,D,V=np.linalg.svd(F_Final)
    V=V.T

    Cxxi=np.linalg.inv(Cxx)
    Cxxi[Cxxi<0]=0
    Cxx_sqrt=np.sqrt(Cxxi)
    wx=np.dot(Cxx_sqrt,U)

    Cyyi=np.linalg.pinv(Cyy)
    Cyyi[Cyyi<0]=0
    Cyy_sqrt=np.sqrt(Cyyi)
    wy=np.dot(Cyy_sqrt,V)
    
    #rth vector pair
    wValues, vVectors = np.linalg.eig(wx)
    index = np.argmax(wValues, axis=0)
    
    l = [0]*r
    for i in range (0,r):
        l[i] = wx[i][index]

     # Finding matrix Dx,Dy
    for i in range(len(l)):
        Dx[i][u]=l[i]
        
    wValuesY, vVectorsY = np.linalg.eig(wy)
    indexY = np.argmax(wValuesY, axis=0)
    l1 = [0]*r
    for i in range (0,r):
        l1[i] = wy[i][indexY]

    for i in range(len(l1)):
        Dy[i][u]=l1[i] 
Dx
```
