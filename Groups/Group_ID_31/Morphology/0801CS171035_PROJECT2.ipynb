##Morphological Feature Extraction of Image
Morphological image processing is a collection of operations related to the shape or morphology of features in an image, such as boundaries, skeletons, etc.We apply a structuring element to an input image and generate an output image.

###Structuring Element
we probe an image with a small shape or template called a structuring element, which defines the region of interest or neighborhood around a pixel.

###Morphological Operation on Image
###Dilation
Dilation is a technique where we expand the image. It adds the number of pixels to the boundaries of objects in an image. The structuring element controls it.Dilation makes objects more visible and fills in small holes in objects.

###dilation(image,shape,*args)
#####Parameter:
####img
input image
####shape
structuring element('square' or 'rectangle')
####*args:
it takes dimension values based on shape i.e.side length for square and length,breadth for rectangle
#####Return
Dilated Image

```python
# Image before dilation
image=cv.imread(r'E:\7th sem all material\Morphological\Improved\star.png')
plt.imshow(image, cmap='Greys_r')
```

```python
#Reading image and applying dilation on it with structuring element of dimension 4*4

dil=cv.imread(r'E:\7th sem all material\Morphological\Improved\star.png')
dil=obj.dilation(dil,'square',4)
plt.imshow(dil, cmap='Greys_r')
```

###Erosion
Erosion erodes away the boundaries of objects. It is used to remove small white noises from the images. Erosion can also be used to detach two connected images.The thickness of object decreases.

###erosion(img,shape,*args)
####Parameter
####img
input image(RGB)
####shape
structuring element('square' or 'rectangle')
####*args
it takes dimensions values based on shape i.e.side length for square and length,breadth for rectangle
####Return
eroded image

```python
#Before erosion
img=cv.imread(r'E:\7th sem all material\Morphological\Improved\dilated.png')
plt.imshow(img, cmap='Greys_r')
```

```python
#after dilation
ero=obj.erosion(img,'square',4)
plt.imshow(ero, cmap='Greys_r')
```

###Opening
The opening operation erodes an image and then dilates the eroded image, using the same structuring element for both operations.
Morphological opening is useful for removing small objects from an image while preserving the shape and size of larger objects in the image.

###opening(img,shape,*args)
####Parameter:
####img
input image(RGB)
####shape
structuring element('square' or 'rectangle')
####*args
dimension of structuring element depending upon shape

```python
#Before applying opening operation on image

img=cv.imread(r'E:\7th sem all material\Morphological\Improved\sNp.png')
plt.imshow(img, cmap='Greys_r')
```

```python
# It is clearly seen that  white mini size holes are removed( also called 'salt and pepper noise')
img=obj.opening(img,'square',3)
plt.imshow(img, cmap='Greys_r')
```

###Closing
The closing operation dilates an image and then erodes the dilated image, using the same structuring element for both operations.Closing is used for smoothening of contour and fusing of narrow breaks.
###closing(img,shape,*args)
####Parameter
####img
input image(RGB)
####shape
structuring element(square or rectangle)
####*args
dimension of structuring element depending upon shape
####Return
Image obtained after performing closing operation

```python
#Image before applying closing operation 

img=cv.imread(r'E:\7th sem all material\Morphological\Improved\close.png')
plt.imshow(img, cmap='Greys_r')
```

```python
# After applying closing operation
# As we can see that minute holes are filled without changing object size
close=obj.closing(img,'square',4)
plt.imshow(close, cmap='Greys_r')
```
